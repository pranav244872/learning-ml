### Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

### Initalizing X,y,w,b

```python
X = np.array([[1],[2],[3],[4],[5],[6]])
y = np.array([0,0,1,1,1,1])
w = np.array([2.5])
b = -5.7
```

### Predict y given X,w,b

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def predict_y(X,w,b):
  m = y.shape[0]
  y_probs = np.zeros(m)
  for i in range(m):
    y_probs[i] = sigmoid(np.dot(w,X[i])+b)
  y_preds = np.where(y_probs >=0.5, 1,0)
  return y_probs, y_preds
```

```python
predict_y(X,w,b)
```

### Loss Function

```python
def loss_function(X,y,w,b,i):
  EPSILON = 1e-8 #to prevent log(0)
  z = np.dot(X[i],w)+b
  a = sigmoid(z)
  loss = -1*y[i]*np.log(a+EPSILON) - (1-y[i])*(np.log(1-a+EPSILON))
  return loss
```

### Cost function

```python
def cost_function(X,y,w,b):
  m = X.shape[0]
  sum = 0
  for i in range(m):
    sum += loss_function(X,y,w,b,i)
  res = sum/m
  return res
cost_function(X,y,w,-5.7)
```
