Have to predict Grade of the student given all these inputs so lets group X = features Y = Output (Grade Class) from training data
But first we need to one-hot encode the Grade class

```python
grade_class_og = df['GradeClass'].to_numpy()
print(grade_class_og)
```

```python
def one_hot_encode(in_array):
  # Get unique values from the in_array
  uniq_vals = np.unique(in_array)
  no_uniq_vals = len(uniq_vals)

  # Make a 0 matrix which is len of in_array * no_uniq_vals
  one_hot_matrix = np.zeros((len(in_array),no_uniq_vals))

  # Make the zero matrix one-hot-encoded
  for i,j in zip(in_array, one_hot_matrix):
    j[int(i)] = 1

  return one_hot_matrix

T = one_hot_encode(grade_class_og)
T_trial = T[:16]
```

```python
class Layer:

  def sigmoid(self,x):
    return 1 / (1 + np.exp(-x))

  # W is a matrix of size D*N where D = no. of nodes and N = no. of input
  # B is the bias of the layer which is a matrix of D*1 = where D = no. of nodes
  def __init__(self,w,b):
    self.w = w
    self.b = b

  # Given an input layer it gives output after multiplying and applying tanh activation function
  def get_output(self,in_matrix):
    output = np.dot(in_matrix,self.w) + self.b
    output = self.sigmoid(output)
    return output

  def get_softmax(self, in_matrix):
    get_softm_of = np.dot(in_matrix,self.w) + self.b
    # Subtract the max value in each row for numerical stability
    exp_output = np.exp(get_softm_of)
    softmax = exp_output / np.sum(exp_output, axis = 1, keepdims = True)
    return softmax
```

```python
def l_matrix(target, prediction):
  loss_matrix = -np.sum(target * np.log(prediction), axis=1, keepdims=True)
  return loss_matrix

def loss_scalar(target, prediction):
  loss_matrix = l_matrix(target, prediction)
  loss = np.mean(loss_matrix)
  return loss

def classification_rate(target, prediction):
  sum = 0
  for i, j in zip(target, prediction):
    if np.argmax(i) == np.argmax(j):
      sum += 1
  c_rate = sum / target.shape[0]
  return c_rate
```

```python
def derivative_w2(z_1, target, predictions):
  return -np.dot(z_1.T, (target - predictions))

def derivative_b2(target, predictions):
  return -np.sum((target-predictions), axis=0)

def derivative_w1(X, z_1, target, predictions, w2):
  dZ = np.dot((target-predictions),w2.T) * z_1 * (1 - z_1)
  ret2 = np.dot(X.T, dZ)
  return -ret2

def derivative_b1(z_1, target, predictions, w2):
  return -np.sum(np.dot((target-predictions),w2.T) * z_1 * (1 - z_1),axis=0)
```

```python
def gradient_step(learning_rate,w1,b1,w2,b2,target,predictions,z_1,X):
  w1 -= learning_rate*derivative_w1(X, z_1, target, predictions, w2)
  b1 -= learning_rate*derivative_b1(z_1, target, predictions, w2)

  w2 -= learning_rate*derivative_w2(z_1, target, predictions)
  b2 -= learning_rate*derivative_b2(target, predictions)

  return w1, b1, w2, b2
```

```python
w1 = np.ones((13,3))*0.1
b1 = np.zeros((3,))
l1 = Layer(w1, b1)
z_1 = l1.get_output(X_trial)

w2 = np.ones((3,5))*0.1
b2 = np.zeros((5,))
l2 = Layer(w2, b2)
p = l2.get_softmax(z_1)


l1.w, l1.b, l2.w, l2.b = gradient_step(0.1, l1.w, l1.b, l2.w, l2.b, T_trial, p, z_1, X_trial)
z_1 = l1.get_output(X_trial)
p = l2.get_softmax(z_1)
```

```python
for i in range(1100):
  l1.w, l1.b, l2.w, l2.b = gradient_step(0.1, l1.w, l1.b, l2.w, l2.b, T_trial, p, z_1, X_trial)
  z_1 = l1.get_output(X_trial)
  p = l2.get_softmax(z_1)
  if (i % 100 == 0):
    print(loss_scalar(T_trial, p))
    print(classification_rate(T_trial, p))
```
